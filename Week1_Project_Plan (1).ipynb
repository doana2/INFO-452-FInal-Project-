## Project Overview

**Problem Statement:**  
Small businesses receive customer reviews but lack tools to quickly:
- Understand overall sentiment (positive/negative/neutral)
- Answer specific questions about their products using company knowledge
- Extract actionable insights from feedback

**Solution:**  
An AI system that combines sentiment analysis with a RAG system to provide intelligent responses about products.

**Target Users:** Small business owners, customer service teams, product managers

---

## Selected Components (2 Required)

### âœ… Component 1: Fine-Tuned Text Model (DistilBERT)
- **Purpose:** Classify customer review sentiment (positive/negative/neutral)
- **Model:** `distilbert-base-uncased` from Hugging Face
- **Training Method:** Transfer learning - fine-tune on review dataset
- **Dataset:** Amazon Product Reviews (publicly available)

### âœ… Component 2: RAG System
- **Purpose:** Answer questions about products using company documentation
- **Embeddings:** `sentence-transformers/all-MiniLM-L6-v2`
- **Vector DB:** ChromaDB (local, free)
- **Documents:** Product manuals, FAQs, return policies (we'll create sample docs)

### ğŸ Bonus Component 3: GUI (Gradio)
- **Purpose:** User-friendly interface for review analysis and Q&A
- **Features:** Text input for reviews, sentiment display, RAG-powered Q&A
