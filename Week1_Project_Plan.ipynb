## Project Overview

**Problem Statement:**  
Electronics retailers and manufacturers receive thousands of customer reviews but lack tools to quickly:
- Understand overall sentiment (positive/negative)
- Answer specific questions about their products using official documentation
- Provide accurate product information to customers

**Solution:**  
An AI system that combines sentiment analysis with a RAG (Retrieval Augmented Generation) system to analyze reviews and provide intelligent responses about products using official manufacturer documentation.

**Target Users:** Electronics retailers, customer service teams, product support specialists, consumers researching products

**Demo Product:** Sony WH-1000XM5 Wireless Noise Canceling Headphones (using official Sony documentation)

---

## Selected Components (2 Required)

### âœ… Component 1: Fine-Tuned Text Model (DistilBERT)
- **Purpose:** Classify customer review sentiment (positive/negative)
- **Model:** `distilbert-base-uncased` from Hugging Face
- **Training Method:** Transfer learning - fine-tune on Amazon product review dataset
- **Dataset:** Amazon Polarity (publicly available, 3.6M reviews)
- **Training Subset:** 500-2000 samples for efficient training

### âœ… Component 2: RAG System
- **Purpose:** Answer questions about products using official manufacturer documentation
- **Embeddings:** `sentence-transformers/all-MiniLM-L6-v2`
- **Vector DB:** ChromaDB (local, lightweight, free)
- **Documents:** Official Sony WH-1000XM5 documentation (specifications, setup guides, warranty, troubleshooting)

### ğŸ Bonus Component 3: GUI (Gradio)
- **Purpose:** User-friendly web interface for review analysis and product Q&A
- **Features:**
  - Sentiment analysis tab with confidence scores
  - Product Q&A tab with source citations
  - Real-time responses
